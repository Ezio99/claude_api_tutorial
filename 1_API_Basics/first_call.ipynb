```python
from anthropic import Anthropic


# Grabs API key from env

client = Anthropic()
model_name = "claude-sonnet-4-6"
```

```python
from anthropic.types import Message 

def send_message(content:str) -> Message:
    return client.messages.create(
        model=model_name,
        max_tokens=1000,
        messages=[
            {
                "role":"user",
                "content":content
            }
        ]
    )
```

```python
send_message(content="Hi! Do you remember what I just said?").content[0].text
```

```python

from anthropic.types import MessageParam

# Claude messages are stateless you will have to maintain chat history and send the entire history for it to be aware of context

def add_user_message(messages: list[MessageParam], content):
    message = {"role": "user", "content": content}
    messages.append(message)


# LLM Generated message
def add_assistant_message(messages: list[MessageParam], content):
    message = {"role": "assistant", "content": content}
    messages.append(message)



def chat(messages: list[MessageParam]) -> str:
    reply = client.messages.create(
        model=model_name,
        max_tokens=1000,
        messages=messages,
    )
    return reply.content[0].text

```

```python
messages = []

add_user_message(messages,"Hi! My name is Vaibhav")


answer = chat(messages)

add_assistant_message(messages,answer)

add_user_message(messages,"What was my name?")

answer = chat(messages)

answer

add_assistant_message(messages,answer)
```

```python
chat_messages = []

while True:

    user_input = input(">")
    print(f"> {user_input}")
    add_user_message(messages=chat_messages, content=user_input)
    answer = chat(messages=chat_messages)
    print(f"---\n {answer}\n---")
    add_assistant_message(messages=chat_messages, content=answer)
```
