```python
from dotenv import load_dotenv

# Loading data from the enviroment variable
load_dotenv()
```

```python
from anthropic import Anthropic


# Grabs API key from env

client = Anthropic()
model_name = "claude-sonnet-4-6"
```

```python
import Notebooks.helper as helper


chat_messages = []

while True:

    user_input = input(">")
    print(f"> {user_input}")
    helper.add_user_message(messages=chat_messages, content=user_input)
    answer = helper.chat(model_name=model_name, client=client, messages=chat_messages)
    print(f"---\n {answer}\n---")
    helper.add_assistant_message(messages=chat_messages, content=answer)
```
