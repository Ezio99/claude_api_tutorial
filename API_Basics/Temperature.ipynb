```python
from dotenv import load_dotenv
import helper

# Loading data from the enviroment variable
load_dotenv()
```

```python
from anthropic import Anthropic


# Grabs API key from env

client = Anthropic()
model_name = "claude-sonnet-4-6"
```

```python
messages = []

helper.add_user_message(
    messages=messages,
    content="Generate 1 sentence movie idea"
    )

# Temp of 0 means more deterministic, chooses the highest probability words
answer = helper.chat(model_name=model_name,client=client,messages=messages,temperature=0.0)

answer
```

```python
# Increasing the temp means more creative responses (evens up the probability of choosing words)
answer = helper.chat(model_name=model_name,client=client,messages=messages,temperature=1.0)

# Keep in mind though that it is non deterministic and you could still get repeated answers based on luck

answer
```
